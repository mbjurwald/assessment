```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('imdb_split_genres.csv')

exploded_df = df.explode("Genre")

# Group by genre and calculate avg imdb rating
genre_avg_rating = exploded_df.groupby('Genre')['IMDB_Rating'].mean()

# Find top three genres with shortes runtime
top_three_shortest_genres = genre_avg_rating.nsmallest(3)

# Make a pie chart
plt.pie(top_three_shortest_genres, labels=top_three_shortest_genres.index, autopct='%1.2f%%', startangle=140)
plt.title('Top Three Genres With Shortes Runtime')
plt.show()
```
