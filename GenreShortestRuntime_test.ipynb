```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming your DataFrame is named df
df["Genre"] = df["Genre"].str.split(", ")

# Explode the genre creating a duplicate entry for each movie but with a different genre
exploded_df = df.explode("Genre")

# Convert 'IMDB_Rating' to numeric data type
exploded_df['IMDB_Rating'] = pd.to_numeric(exploded_df['IMDB_Rating'], errors='coerce')

# Group by genre and calculate the average IMDb rating
genre_avg_rating = exploded_df.groupby('Genre')['IMDB_Rating'].mean()

# Find the genre with the lowest average IMDb rating
lowest_avg_genres = genre_avg_rating.idxmin()
lowest_avg_rating = genre_avg_rating.min()

# Plot the bar chart
plt.bar(lowest_avg_genres, lowest_avg_rating)

# Add labels and title
plt.xlabel('Genre')
plt.ylabel('Average IMDb Rating')
plt.title('Genre with the Lowest Average IMDb Rating')

# Show the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Antag att din DataFrame heter df
df["Genre"] = df["Genre"].str.split(", ")

# Explodera genrerna för att skapa en duplicerad post för varje film med en annan genre
exploded_df = df.explode("Genre")

# Konvertera 'IMDB_Rating' till numerisk datatyp
exploded_df['IMDB_Rating'] = pd.to_numeric(exploded_df['IMDB_Rating'], errors='coerce')

# Gruppera efter genre och beräkna det genomsnittliga IMDb-betyget
genre_avg_rating = exploded_df.groupby('Genre')['IMDB_Rating'].mean()

# Hämta de tre genrer med kortast medelruntid
top_three_shortest_genres = genre_avg_rating.nsmallest(3)

# Skapa ett cirkeldiagram
plt.pie(top_three_shortest_genres, labels=top_three_shortest_genres.index, autopct='%1.2f%%', startangle=140)

# Lägg till titel
plt.title('Tre Genrer med Kortast Medelruntid')

# Visa cirkeldiagrammet
plt.show()
```
