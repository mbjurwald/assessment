```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

df = pd.read_csv('imdb_top_1000.csv', index_col=0)

df["Gross"] = df["Gross"].str.replace(",", "")  # Ersätt alla kommatecken med ingenting, dvs. ta bort dem
df["Gross"] = df["Gross"].astype(float)  # Konvertera strängar till flyttal med `astype()`

# Filtrera filmer av Alfred Hitchcock
ah_movies = df[df['Director'] == 'Alfred Hitchcock']

top_three_movies = ah_movies.head(3)

# Plot the bar chart
plt.bar(top_three_movies['Series_Title'], top_three_movies['Gross'])

# Add labels and title
plt.xlabel('Series Title')
plt.ylabel('Gross')
plt.title('Alfred Hitchcock top five Gross')

# Show the plot
plt.show()
```
